```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
g = sns.barplot(data=titanic, 
                x = 'sex', 
                y = 'survived', 
                # order = [ColumnName],
                # hue_order = [ColumnName]
                hue = 'pclass')

g.set(
    xlabel ="gender",
    ylabel ='survived',
    title = "Comparison"
    )
```

```python
plt.pie((titanic['survived'].value_counts()), labels= ['Not Survived','That Lived'], autopct='%.2f%%', startangle=90, colors = sns.color_palette('dark'))
```

```python
# colors =  ['red','green'] 
labels= ['Drowned','That Lived']
colors = sns.color_palette("pastel")[0:2]
percentage='%.1f%%'
wedgeprops={"width": 0.6} # donut chart

plt.pie(titanic['survived'].value_counts(), labels=labels, colors = colors, autopct=percentage, startangle=90, wedgeprops= wedgeprops)
```
